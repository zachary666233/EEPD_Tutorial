```python
"""情感分析"""
#停用词问题#
positive_words=['好看','美丽','漂亮']
negative_words=['丑','难看','脏']
sentence='这个公园好难看啊，难看！但是水非常美丽'
def senti_score(sentence):
    '''基于词典和规则的情感打分'''
    pos=0
    neg=0
    word_list=jieba.lcut(sentence,cut_all=False)
    print(word_list)
    for word in word_list:
        if word in positive_words:
            pos+=1
        elif word in negative_words:
            neg+=1
        else:
            continue
    return pos,neg
senti_score(sentence)
```

```python
#ASLA2018 student award爬虫#

def get_link():
    '''获取一年度所有奖项'''
    url='https://www.asla.org/2018studentawards/index.html'
    html = requests.get(url).content
    soup = BeautifulSoup(html, 'lxml')
    locations = soup.find_all('h3')
    links=[]
    for location in locations:
        link='https://www.asla.org/2018studentawards/'+str(location.a['href'])
        links.append(link)
    return links

def get_pic(url):
    '''单个奖项获取图片'''
    html=requests.get(url).content
    soup=BeautifulSoup(html,'lxml')
    pics=soup.find_all('div',class_='item')
    pic_total=[]
    for pic in pics:
        link='https://www.asla.org/2018studentawards/'+str(pic.img.attrs['src'])
        #print(link)
        pic_total.append(link)
    return pic_total

def main():
    links=get_link()
    for link in links:
        imgs=get_pic(link)
        #print(imgs)
        for img in imgs:
            print(img)
#             with open(str(random.choice([i for i in range(0,1500)]))+'.jpg','wb') as image:
#                 image.write(requests.get(img).content)


main()
```

```python
"""机器学习"""
#https://blog.csdn.net/little_fire/article/details/81062447#鸢尾花数据集
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
iris = sklearn.datasets.load_iris()
# print(iris)
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25, 
                                                    random_state=10)
# print(X_train)

model=KNeighborsClassifier(n_neighbors=5)
"""KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',
           metric_params=None, n_jobs=None, n_neighbors=5, p=2,
           weights='uniform')"""
model.fit(X_train,y_train)
y_predict=model.predict(X_test)
# print(y_predict)
metric=sklearn.metrics.confusion_matrix(y_predict,y_test)
auc=sklearn.metrics.accuracy_score(y_predict,y_test)
print(auc)
plt.imshow(metric, interpolation='nearest')
plt.colorbar()
plt.show()
```
